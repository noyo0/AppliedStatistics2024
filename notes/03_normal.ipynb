## Flipping Several Coins

```python
# Flip a single coin.
np.random.binomial(1, 0.5)
```

```python
# Flip a single coin several times.

#     random samples from a binomial distribution.   
flips = np.random.binomial(
    1, # 1 trial per sample 
    0.5, # probability of success (head) per trial
    1000 # number of samples
    )

# Show outcomes.
flips
```

```python
# Count heads.
heads = flips.sum()

# Show.
heads
```

```python
# Create a blank plot.
fig, ax = plt.subplots()

# Create a bar chart.
ax.bar([0, 1], [1000 - heads, heads])

# Add x labels.
ax.set_xticks([0, 1])

# Add title.
ax.set_title('Flipping a coin 1000 times')

# Add axis label.
ax.set_xlabel('No heads')
ax.set_ylabel('Count');
```

```python
# Flip two coins several times.
twoflips = np.random.binomial(
    2, # 2 coinflips per trial
    0.5, #50% probability
    1000 #samples
    )

# Show how many times I got heads per sample 
# 0,0 means no heads, 0,1 and 1,0 means head and tail, 1,1 is heads only 
twoflips
```

```python
# Count the number of times each value occurs.
uniques, counts = np.unique(twoflips, return_counts=True)

# Show.
uniques, counts
```

```python
# Create a blank plot.
fig, ax = plt.subplots()

# Create a bar chart.
ax.bar(uniques, counts)

# Set x ticks.
ax.set_xticks(uniques)

# Add title.
ax.set_title('Flipping two coins 1000 times')

# Add axis label.
ax.set_xlabel('Number of heads')
ax.set_ylabel('Count')
```

## Probability Density Function

$ f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $

https://en.wikipedia.org/wiki/Normal_distribution
>The function \( f(x) \) computes the value of the normal distribution for a given \($ x $\), centered at \( $\mu$ \)(mu) with a spread defined by \( $\sigma $\) (*sigma*). The output of this function is the height of the Gaussian curve at the point \( x \), which reflects the likelihood of observing that value in a normally distributed dataset.


```python
# Define the probability density function (PDF) of a Gaussian (normal) distribution.------
def f(x, mu=0.0, sigma=1.0):
    # The normalization constant for the Gaussian distribution.
    # This ensures that the total area under the curve equals 1.
    norm = 1.0 / np.sqrt(2.0 * np.pi * sigma**2) # This line calculates the normalization constant for the Gaussian distribution
    
    # Calculate the value of the normal distribution at point x.
    # The expression inside the exponential determines how far x is from the mean (mu),
    # scaled by the variance (sigma squared).
    return norm * np.exp(-(x - mu)**2 / (2.0 * sigma**2)) # The function computes the value of the normal distribution for the input x
```

```python
# Test the function.
f(0.0)
```

To understand why `f(0.0)` outputs `0.3989422804014327`, let's break down the calculation step by step, assuming default values for the parameters \( $\mu$ \) (mean) and \( $\sigma$ \) *sigma* (standard deviation):

### Parameters Used
- \( $x = 0.0$ \)
- \( $\mu = 0.0$ \) (default value)
- \( $\sigma = 1.0$ \) (default value)

### Step-by-Step Calculation

1. **Calculate the Normalization Constant:**
   \[
   $\text{norm} = \frac{1}{\sqrt{2 \pi \sigma^2}} = \frac{1}{\sqrt{2 \pi (1.0)^2}} = \frac{1}{\sqrt{2 \pi}} \approx 0.3989422804014337$
   \]

2. **Calculate the Exponential Component:**
   \[
   $\text{Exponent} = -\frac{(x - \mu)^2}{2 \sigma^2} = -\frac{(0.0 - 0.0)^2}{2 (1.0)^2} = -\frac{0}{2} = 0$
   \]

3. **Compute the Final Value:**
   \[
   $f(0.0) = \text{norm} \cdot \exp(0) = \text{norm} \cdot 1 = 0.3989422804014337 \cdot 1 = 0.3989422804014337$
   \]

### Summary
The output `0.3989422804014327` corresponds to the height of the Gaussian (normal) distribution curve at the point \( x = 0.0 \) when the mean \( \mu \) is 0 and the standard deviation \( \sigma \) is 1. This value represents the likelihood of observing a value of \( 0.0 \) in a standard normal distribution.


```python
# Create an x range representing the number of standard deviations
    # from the mean (mu = 0) in a normal distribution. 
    # This range goes from -5 to 5 with 1001 points.
x = np.linspace(-5, 5, 1001)

# Get y values by evaluating the normal distribution function f(x)
    # at each point in the x range. This will provide the probability
    # density for each value of x, which can represent outcomes of coin flips.
y = f(x)

# The x values represent the range of deviations from the mean,
# and the y values represent the corresponding probabilities for those outcomes.
x, y
```

```python
# Create an empty plot.
fig, ax = plt.subplots()
# Plot the function.
ax.plot(x, y)
```

Reference: [6.4: Normal Approximation to the Binomial Distribution](https://stats.libretexts.org/Courses/Las_Positas_College/Math_40%3A_Statistics_and_Probability/06%3A_Continuous_Random_Variables_and_the_Normal_Distribution/6.04%3A_Normal_Approximation_to_the_Binomial_Distribution)

```python
# Number of coins to flip.
n = 100  # Set the number of coins to be flipped in each trial.

# Number of times to do it.
k = 10000 * n  # Total number of trials to perform (10000 times flipping n coins).

# Probability of heads.
p = 0.5  # Probability of getting heads for each coin flip (50%).

# Flip two coins several times.
nflips = np.random.binomial(n, p, k)  # Perform k trials of flipping n coins and record the number of heads.

# Count the number of times each value occurs.
uniques, counts = np.unique(nflips, return_counts=True)  # Get unique counts of heads and their frequencies.

# Create a blank plot.
fig, ax = plt.subplots(figsize=(12, 5))  # Initialize a new figure with specified size for the plot.

# Create a bar chart.
ax.bar(uniques, counts / k)  # Create a bar chart showing the proportion of each outcome (number of heads).

# Create an x range for plotting the normal distribution.
x = np.linspace(25, 75, 1001)  # Generate 1001 evenly spaced values from 25 to 75 for the x-axis.

# Get y values for the normal distribution.
y = f(x, mu=n*p, sigma=np.sqrt(n * p * (1 - p)))  # Compute the y values using the normal distribution function f(x).

# Plot the normal distribution function.
ax.plot(x, y, 'r-')  # Overlay the normal distribution curve on the bar chart in red.

# Set x ticks for the bar chart.
ax.set_xticks(uniques)  # Set the x-axis ticks to correspond to the unique values of heads.

# Add title and labels
ax.set_title(f'Flipping {n} coins {k} times')  
ax.set_xlabel('No heads')  # Label for the x-axis (number of heads).
ax.set_ylabel('Count')  # Label for the y-axis (count of occurrences).
```

#### Probability PLot (QQ Plot)

A QQ plot (Quantile-Quantile plot) is a graphical tool used to assess if a dataset follows a specified distribution, typically a normal distribution.

<br>Key Features:

- **X-axis**: Theoretical quantiles from the specified distribution (e.g., normal).
- **Y-axis**: Sample quantiles from the dataset.
- **Line**: If the data points fall approximately along a straight line (45-degree line), it suggests that the data follows the specified distribution.

<br> Usage:

- **Normality Check**: Visually assess if a dataset follows a normal distribution, which is crucial in many statistical analyses.
- **Residual Analysis**: Evaluate the normality of residuals in regression models to validate assumptions.
- **Comparing Distributions**: Compare two datasets to see if they come from the same distribution, helping in hypothesis testing.
- **Outlier Detection**: Identify potential outliers by observing points that deviate significantly from the reference line.


###### QQ Plots Visualisation (https://library.virginia.edu/data/articles/understanding-q-q-plots)<br> statsmodels.qqplot()  https://www.statsmodels.org/dev/generated/statsmodels.graphics.gofplots.qqplot.html



```python
#import statsmodels as sm
#stats.probplot(data, dist='norm', plot=ax)

import scipy.stats as stats

stats.probplot(data, dist="norm", plot=plt)
plt.title("QQ Plot - Probability Plot")
plt.show()
```

In a QQ plot:
- The red line represents the expected quantiles of the assumed normal distribution.
- The blue dots (or points) represent the actual quantiles of your dataset.
>Interpretation:<br>
If the blue dots closely follow the red line, it suggests that the dataset is approximately normally distributed.<br>
If the dots deviate significantly from the line, especially at the ends, it indicates that the data may not follow a normal distribution.

## Shapiro-Wilk Test

https://en.wikipedia.org/wiki/Anscombe%27s_quartet
https://www.research.autodesk.com/publications/same-stats-different-graphs/#:~:text=Developed%20by%20F.J.%20Anscombe%20in,the%20datasets%20are%20quire%20similar.

The Anscombe's quartet comprises four datasets that have nearly identical simple descriptive statistics, yet have very different distributions and appear very different when graphed. <br>
The four datasets composing Anscombe's quartet. All four sets have identical statistical parameters, but the graphs show them to be considerably different
<br>Anscombe's Quartet illustrates that relying solely on summary statistics (like mean and variance) can be misleading. Each dataset has the same statistics but different distributions.
<br>**Shapiro-Wilk Test** can be used to assess normality for each dataset in Anscombes Quartet. By applying the test, you can reveal that some datasets may not follow a normal distribution despite similar statistics.
>Shapiro-Wilk Test is a statistical test used to determine if a dataset is normally distributed. It tests the null hypothesis that the data comes from a normal distribution. A low p-value (typically < 0.05) indicates that the data is not normally distributed.

```python
# perform Shapiro-Wilk test
import scipy.stats as stats
stats.shapiro(data)
```

The output of the Shapiro-Wilk:
- Statistic: 0.9974806308746338 (This value is close to 1, which suggests that the data is likely normally distributed.)
- P-value: 0.12574808299541473 (A p-value that is much greater than the common significance level (e.g., 0.05).)
>Conclusion:<br>
The test failed to reject the null hypothesis, meaning there is no significant evidence to suggest that the data is *not* normally distributed. **The data can be considered normally distributed based on this test.**


## Central Limit Theorem (CLT)

The **Central Limit Theorem (CLT)** states that, given a sufficiently large sample size, the sampling distribution of the sample mean will be approximately normally distributed, regardless of the original population distribution.

### Key Points:
1. **Sample Size**: A sample size of 30 or more is typically considered sufficient for the theorem to hold.
2. **Independence**: The samples must be independent of each other.
3. **Mean and Variance**: 
   - The mean of the sampling distribution will equal the population mean.
   - The variance will be the population variance divided by the sample size.

### Importance:
- The CLT allows for the use of normal distribution approximations in inferential statistics, making it easier to conduct hypothesis tests and create confidence intervals, even if the population distribution is not normal.
