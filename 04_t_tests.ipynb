## Imports

```python
# Generate data

# Group A
group_a = np.random.normal(0.07,0.02,30)

# Group B
group_b = np.random.normal(0.03,0.02,30)
```

```python
sns.stripplot(data=[group_a,group_b])
sns.boxplot(data=[group_a,group_b], boxprops=dict(alpha=0.5))
```

## Samples and Populations

# Independent two-sample $t$-test
Equal sample sizes and variance
- Given two groups (1, 2), this test is only applicable when:
  - the two sample sizes are equal,
  - it can be assumed that the two distributions have the same variance.
   
The t statistic to test whether the means are different can be calculated as follows:

$$
t = \frac{\bar{X}_1 - \bar{X}_2}{s_p \sqrt{\frac{2}{n}}}
$$

where

$$
s_p = \sqrt{\frac{s_{X_1}^2 + s_{X_2}^2}{2}}
$$

```python
# perform a t-test
stats.ttest_ind(group_a,group_b)
```

## Paired Sample $t$-test

```python
# A group of normally distributed values
group_a = np.random.normal(5,0.2,30)

# Add group_b with normally distributed "noise" values
group_b = group_a + np.random.normal(0.0,0.1,30)
```

```python
# Hisatogram + Shapiro test to cehck normal distribution
plt.hist(group_a, color="blue",alpha=0.5, label="Group A")
plt.hist(group_b, color="green",alpha=0.5, label="Group B")
plt.show()
import scipy.stats as stats
stats.shapiro(group_a)
```

```python
# strip plot
sns.stripplot(data=[group_a,group_b])
```

```python
# Paired sample t-test (https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html#ttest-rel)
stats.ttest_rel(group_a,group_b)
```

- **Statistic**: 1.19  Moderate difference between groups.
- **p-value**: 0.2433  No significant difference (p > 0.05).
- **df**: 29  Degrees of freedom.

**Conclusion**: No significant difference between the groups.
